**Introduction:** A heart attack is a myocardial infarction (tissue death in the muscle layer of the heart) that happens when blood clots occur that block the blood flow to heart muscles, causing them unable to receive enough oxygen. According to the World Health Organization (WHO) statistics, an estimated 17.9 million people die from Cardiovascular Disease each year, accounting for 32% of all global deaths. Among these deaths, 85% were due to heart attack and stroke. Analysis of the factors involved in causing heart attacks is important to be able to assess individuals at risk and detect the possibility of a heart attack before it happens. Furthermore, understanding the common attributes leading to heart disease and attacks is a key factor in prevention, when that information is used to educate and enforce awareness.

To investigate the impacts of different factors including age, gender and lifestyles on heart attack prevalence, data collection  from 303 heart attack patients will be studied and analyzed to answer the following questions. After what age is an individual more likely to develop a heart attack? What is the threshold for healthy levels of cholesterol and at what amount does it become a problem and have adverse effects on heart health? This data set recorded the information of heart attack patients in 14 columns of age, gender, type of chest pain, Cholesterol level, resting blood pressure (in mm Hg), fasting blood sugar > 120 mg/dl, resting electrocardiographic results, maximum heart rate achieved by the patients, exercise-induced angina and their previous peak (ST depression induced by exercise relative to rest).


**Preliminary exploratory data analysis:** This data set contained all comma-separated values and is UTF-8 encoded, uploading action to Jupyter Notebook was successful.


```r
library(tidyverse)
heart <- read_csv("heart.csv") %>%
mutate(sex = as_factor(sex))
heart_plot <- heart %>%
ggplot(aes(x = chol, y = age)) +
geom_point(aes(color = cp)) +
labs(x = "Cholesterol Level of Patients",
    y = "Age",
    color = "Chest Pain Type") +
ggtitle("Patient Chest Pain Type in relation to Age and Cholesterol")
heart_plot
```

```r
heart_data <- select(heart,
      chol, cp, age)
heart_data
nrow(heart_data)
summary(heart_data)
```

**Methods:** We will likely omit columns including the previous peak, resting electrocardiographic results, and fasting blood sugar. Using the data columns age and cholesterol level and the trends observed, we will build a k-nearest neighbours classification algorithm that characterizes the type of chest pain an individual is likely to experience as normal, asymptomatic, angina or abnormal. We specifically wanted to examine cholesterol levels in relation to age in order to stress the importance of monitoring cholesterol and possible chest pains with an increase in age. 
Our hypothetical graph is a scatter plot performed by Cholesterol levels in the x-axis and age on the y-axis, with four different colours to represent different types of chest pain (as shown on the left).


**Expected outcomes and significance:** Previous knowledge had suggested that Cholesterol can cause the development of fatty deposits in blood vessels which eventually leads to the blockage of blood flow within our body. Therefore, we expect to find individuals who are older or who have higher levels of cholesterol in their blood to be at greater risk for chest pain. We also expect a sex bias towards males. These findings will help us define the factors that induce various types of chest pains and the corresponding countermeasures that can be taken to limit major chest pain. Furthermore, this investigation is important for future studies of other types of heart diseases with similar symptoms or pathogens as heart attacks.
